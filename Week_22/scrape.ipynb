```python
from requests import get
from bs4 import BeautifulSoup
import re
from warnings import warn
from time import time
from IPython.core.display import clear_output
import numpy as np
import pandas as pd
```

```python
response = get('https://boston.craigslist.org/search/apa?hasPic=1&max_price=2000&availabilityMode=0')
soup = BeautifulSoup(response.text, 'html.parser')

postings = soup.find_all('li', class_= 'result-row')
print(type(postings)) 
print(len(postings))
```

```python
timing = []
neighborhood = []
title_texts = []
bedroom_counts = []
sqfts = []
links = []
prices = []
```

```python
for i in range(50):
    listing = postings[i]
    if listing.find('span', class_ = 'result-hood') is not None:

            listing_datetime = listing.find('time', class_= 'result-date')['datetime']
            timing.append(listing_datetime)


            listing_hood = listing.find('span', class_= 'result-hood').text
            neighborhood.append(listing_hood)


            listing_title = listing.find('a', class_='result-title hdrlnk')
            listing_title_text = listing_title.text
            title_texts.append(listing_title_text)


            listing_link = listing_title['href']
            links.append(listing_link)
            

            listing_price = int(listing.a.text.strip().replace(",","").replace("$", "")) 
            prices.append(listing_price)
            
            if listing.find('span', class_ = 'housing') is not None:
                
                if 'ft2' in listing.find('span', class_ = 'housing').text.split()[0]:
                    
                    bedroom_count = np.nan
                    bedroom_counts.append(bedroom_count)
                    

                    sqft = int(listing.find('span', class_ = 'housing').text.split()[0][:-3])
                    sqfts.append(sqft)
                    

                elif len(listing.find('span', class_ = 'housing').text.split()) > 2:
                    

                    bedroom_count = listing.find('span', class_ = 'housing').text.replace("br", "").split()[0]
                    bedroom_counts.append(bedroom_count)
                    

                    sqft = int(listing.find('span', class_ = 'housing').text.split()[2][:-3])
                    sqfts.append(sqft)
                    

                elif len(listing.find('span', class_ = 'housing').text.split()) == 2:
                    

                    bedroom_count = listing.find('span', class_ = 'housing').text.replace("br", "").split()[0]
                    bedroom_counts.append(bedroom_count)
                    

                    sqft = np.nan
                    sqfts.append(sqft)                    
                
                else:
                    bedroom_count = np.nan
                    bedroom_counts.append(bedroom_count)
                
                    sqft = np.nan
                    sqfts.append(sqft)
                
            else:
                bedroom_count = np.nan
                bedroom_counts.append(bedroom_count)
                
                sqft = np.nan
                sqfts.append(sqft)
```

```python
boston_apts = pd.DataFrame({'listed': timing,
                        'neighborhood': neighborhood,
                       'list title': title_texts,
                       'number bedrooms': bedroom_counts,
                        'price': prices,
                        'sqft': sqfts,
                        'URL': links
                       })
print(boston_apts.info())
boston_apts.head(10)
```
