## 1a. In a jupyter notebook markdown cell, describe the difference between linear and rank correlation.

- Linear correlation - a measure of the degree of co-relation between two variables that are assumed to have a linear relationship

- Rank correlation coefficient - measures the degree of similarity between two rankings which is used to assess the significance of the relation between the variables

## 1b. Create a scatter plot showing the relationship between calculated_host_listings_count and availability_365 in the Belize Airbnb data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
belize_df = pd.read_csv("belize_airbnb_listings.csv")
```

```python
sns.scatterplot(x='calculated_host_listings_count',y='availability_365',data=belize_df)
plt.suptitle("Calculated Host Listings Count by Availability (365)")
```

## 1c. Calculate the correlation coefficient for the data in the scatter plot, and describe what is happening

```python
np.corrcoef(belize_df['calculated_host_listings_count'],belize_df['availability_365'])
```

The correlation coefficient is 0.17275505.

There is weak positive correlation between 'calculated_host_listings_count' and 'availability_365'

## 2. Create a box and whisker plot for the Belize Airbnb data price. In a markdown cell, describe what is going on in as simple of terms as possible.

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax=sns.boxplot(data=belize_df[["price"]], orient='h')
plt.title("Price distribution")
plt.show()
```

- Price distribution is pretty wide as it ranged from 20 to 11788.

- The box plot is right skewed since there are many outliers above the max.

- Most prices are closer around the left of the median compared to prices right of the median
