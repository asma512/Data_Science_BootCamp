```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].value_counts())
players = pd.DataFrame({"Total players": [player_count]})
players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = len(purchase_data["Item ID"].unique())

average_price = purchase_data["Price"].mean()

number_of_purchase = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()

purchasing_analysis_total = pd.DataFrame({"Number of Unique Items":[unique_item],
                                   "Average Price": average_price,
                                   "Number of Purchases": number_of_purchase,
                                   "Total Revenue": total_revenue})

format_mapping = {"Average Price": "${:,.2f}","Total Revenue": "${:,.2f}"}
purchasing_analysis_total.style.format(format_mapping)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = purchase_data.groupby("Gender")

total_count_gender= gender_group["SN"].nunique()

percent_player = round((total_count_gender/player_count)*100,2)


gender_demo=pd.DataFrame({"Total Count":total_count_gender, "Pecentage of Players": percent_player})
gender_demo.index.name=None
gender_demo.sort_values(["Total Count"], ascending = False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = gender_group["Purchase ID"].count()
average_purchase_price = gender_group["Price"].mean()
total_purchase_value = gender_group["Price"].sum()
avg_total_purchase_per_person = total_purchase_value/total_count_gender

gender_purchases=pd.DataFrame({"Purchase Count":purchase_count, 
                           "Average Purchase Price": average_purchase_price,
                          "Total Purchase Value": total_purchase_value,
                          "Avg total purchase per person":avg_total_purchase_per_person})

format_mapping = {"Average Purchase Price": "${:,.2f}","Total Purchase Value": "${:,.2f}",
                  "Avg total purchase per person":"${:,.2f}"}
gender_purchases.style.format(format_mapping)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 122]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=age_labels)


age_groups = purchase_data.groupby("Age Group")
total_count_age = age_groups["SN"].nunique()
percentage_of_players = round((total_count_age/player_count) *100,2)


age_demo= pd.DataFrame({"Total Count":total_count_age, "Percentage of Players":percentage_of_players})
age_demo.index.name=None
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count=age_groups["Purchase ID"].count()
average_purchase_price = age_groups["Price"].mean()
total_purchase_value = age_groups["Price"].sum()

avg_total_purchase_per_person = total_purchase_value/total_count_age

age_purchases =pd.DataFrame({"Purchase Count":purchase_count, 
                           "Average Purchase Price": average_purchase_price,
                          "Total Purchase Value": total_purchase_value,
                          "Avg total purchase per person":avg_total_purchase_per_person})

format_mapping = {"Average Purchase Price": "${:,.2f}","Total Purchase Value": "${:,.2f}",
                  "Avg total purchase per person":"${:,.2f}"}
age_purchases.style.format(format_mapping)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_group= purchase_data.groupby("SN")

purchase_count = spender_group["Purchase ID"].count()

average_purchase_price = spender_group["Price"].mean()

total_purchase_value = spender_group["Price"].sum()

spenders = pd.DataFrame({'Purchase Count': purchase_count,
                             'Average Purchase Price': average_purchase_price,
                             'Total Purchase Value': total_purchase_value})

#spenders.sort_values('Total Purchase Value',ascending=False).head(5)

top_spenders=spenders.sort_values('Total Purchase Value',ascending=False).head(5)
format_mapping = {"Average Purchase Price": "${:,.2f}","Total Purchase Value": "${:,.2f}"}
top_spenders.style.format(format_mapping)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_group= purchase_data.groupby(["Item ID","Item Name"])

purchase_count =items_group["Purchase ID"].count()

average_purchase_price = items_group["Price"].mean()

total_purchase_value = items_group["Price"].sum()


popular_items = pd.DataFrame({'Purchase Count': purchase_count,
                             'Average Purchase Price': average_purchase_price,
                             'Total Purchase Value': total_purchase_value})

#popular_items.sort_values('Purchase Count',ascending=False).head(5)

top_items = popular_items.sort_values('Purchase Count',ascending=False).head(5)

format_mapping = {"Average Purchase Price": "${:,.2f}","Total Purchase Value": "${:,.2f}"}
top_items.style.format(format_mapping)
```
