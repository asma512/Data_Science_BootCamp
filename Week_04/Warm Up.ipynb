```python
import pandas as pd
```

```python
belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df
```

```python
belize_df.columns
```

```python
host_group=belize_df.groupby(["name"])

count = host_group['calculated_host_listings_count'].count()

price=host_group["price"].min()

df = pd.DataFrame({'Count': count,
                    'Price': price})
df.sample(50)
```
