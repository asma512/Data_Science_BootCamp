### a. Import the data into a pandas dataframe

```python
df = pd.read_csv("movies.csv",parse_dates=True,header=1, index_col='Week')
df.rename(columns={"movies: (United States)": "count"},inplace=True)
df.tail()
```

```python
plt.plot(df.index, df['count'])
plt.xlabel('Time')
plt.ylabel('Search Count')
plt.title('Movies search count')
plt.show()
```

### b. Decide whether ARMA, ARIMA, or SARIMA makes more sense for the dataset

```python
df['PCT_Change']=df['count'].pct_change()
df['PCT_Change'].plot()
plt.xlabel('time')
plt.ylabel('PCT Change')
plt.title('Movies PCT Change')
plt.show()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
plot_acf(df[['PCT_Change']], lags = 10, missing='drop');
```

```python
plot_pacf(df[['PCT_Change']], zero = False, lags = 10);
```

I think ARMA makes the least sense since the orginal data was not depend on the time at which the series is observed.

I think SARIMA makes most sense since it accounts for seasonality reflected in the data.

## c. Run either an ARMA, ARIMA, or SARIMA on it

```python
# specify training data
data = df['PCT_Change']
# define model configuration
my_order = (1, 1, 1)
my_seasonal_order = (1, 1, 1, 12)
# define model
model = SARIMAX(data, order=my_order, seasonal_order=my_seasonal_order)
results = model.fit()
results.summary()
```

```python
results_df = pd.DataFrame(results.forecast(steps=15))
results_df.plot(title = '25 week Forecast')
```

We see the number of searches for movies increase in the end of december then drastically for Januarary. 
We see this pattern repeating for the following months 

## 2. What is a fourier transform? Perform one on y = np.cos(x/2), and both show (with code) and explain the key elements.

The Fourier Transform is a mathematical operation that changes the domain (x-axis) of a signal from time to frequency.
It takes time-based pattern, measures every possible cycle, and returns the overall elements such as the amplitude, offset, & rotation speed for every cycle that was found.

```python
x = np.arange(1,120,1)
y = np.cos(x/2)
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.title('y=cos(x/2)')
plt.show()
```

```python
#Power Spectral Density (PSD)
psd = np.abs(np.fft.rfft(y))
#Frequency
freqs = np.fft.rfftfreq(len(y))
plt.plot(freqs,psd)

plt.axvline(1/(4*np.pi),c="orange")
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.title('Fourier Transformation')
plt.show()
```
