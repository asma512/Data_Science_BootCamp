```python
#Starter code. Do NOT modify this cell
import pandas as pd
import numpy as np

# Read in dataset
df = pd.read_csv('cars.csv',delimiter=';')

# Convert each column to an np.array
car = df.Car.to_numpy()
mpg = df.MPG.to_numpy()
cylinders = df.Cylinders.to_numpy()
displacement = df.Displacement.to_numpy()
horsepower = df.Horsepower.to_numpy()
weight = df.Weight.to_numpy()
acceleration = df.Acceleration.to_numpy()
model = df.Model.to_numpy()
origin = df.Origin.to_numpy()

# create an array of headers
headers = np.array(['Car', 'MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight','Acceleration', 'Model', 'Origin'])
```

# The Unix Shell

- A unix shell is a command-line interpreter. The shell provides a command line interface for the users to interact with the unix-like operating system.
- Some commands are 'ls' to list files, 'cp' to copy files, and 'man' to get more information on commands
- There are different Unix shells, most notably bash, sh, C Shell and zsh. 
- The Bourne Shell (sh) was the first shell programs that came with Unix.
